# Orientação a Objetos

- Objetos

  Um objeto é uma instância dessa classe.
Ele é criado a partir da classe e representa algo tangível do mundo real.
Por exemplo, se criarmos um objeto “MeuCarro” a partir da classe “Carro”, ele terá as características e comportamentos definidos pela classe.
- Classe

  Uma classe é como um modelo ou planta baixa que define as características (propriedades) e comportamentos (métodos) que os objetos dessa classe possuirão1.
Ela serve como um molde para criar objetos.
Por exemplo, se tivermos uma classe chamada “Carro”, ela pode definir propriedades como “cor”, “marca” e métodos como “ligar” e “desligar”.
- Método

  Os métodos são ações que um objeto pode realizar.
Eles são definidos na classe e podem ser chamados nos objetos.
No exemplo do carro, os métodos “ligar” e “desligar” seriam ações que o objeto “MeuCarro” pode executar.
- Herença

  A herança é um mecanismo que permite criar uma nova classe com base em uma classe existente.
A classe filha herda as propriedades (atributos) e comportamentos (métodos) da classe pai.
Isso promove a reutilização de código e a organização hierárquica das classes.
Por exemplo, se tivermos uma classe “Animal” com métodos como “respirar” e “mover”, podemos criar classes filhas como “Cachorro” e “Gato” que herdam esses métodos.
- Polimorfismo

  O polimorfismo permite que objetos de diferentes classes sejam tratados de maneira uniforme.
Isso significa que podemos usar um método com o mesmo nome em várias classes diferentes.
Por exemplo, tanto um objeto “Cachorro” quanto um objeto “Gato” podem ter um método “emitirSom”, mas cada um deles implementa esse método de maneira específica

- Encapsulamento

  O encapsulamento refere-se à ideia de esconder detalhes de implementação e expor apenas uma interface pública.
Isso é alcançado usando métodos de acesso (getters e setters) para acessar ou modificar os atributos de um objeto.
Por exemplo, podemos ter um atributo “idade” em uma classe “Pessoa”, mas o acesso a esse atributo é feito por meio de métodos como “getIdade” e “setIdade”
  

# Estatística

- Média

  Média (Média Aritmética):

A média é calculada somando todos os valores de um conjunto de dados e dividindo pelo número de elementos.
Ela representa um valor central do conjunto.
Fórmula: Me​=nx1​+x2​+x3​+…+xn​​

É útil para sintetizar informações e facilitar a interpretação dos dados12.



Vamos a exemplos:


Média de Idade:

Jogadores de basquete têm as idades: 28, 27, 19, 23 e 21 anos.
Média de idade: Me​=528+27+19+23+21​=23.6
 anos.



Média de Altura:

Alturas de alunos: 1,54 m; 1,67 m; 1,50 m; 1,65 m; 1,75 m; 1,69 m; 1,60 m; 1,55 m; 1,78 m.
Média de altura: Me​=1,65m

- Mediana

  A mediana é um valor central em um conjunto de dados ordenado. Ela divide o conjunto em duas partes com a mesma quantidade de elementos1. Aqui estão os passos para calcular a mediana:

Organize os dados em ordem crescente ou decrescente.
Verifique o número de elementos no conjunto:

Se for ímpar, a mediana é o elemento na posição 2n+1​
.
Se for par, a mediana é a média aritmética entre os elementos nas posições 2n​
 e 2n​+1
.



Exemplo 1 (idades de estudantes):

Idades: 11, 11, 11, 11, 12, 12, 12, 13, 13.
Como o número de elementos é ímpar (9), a mediana é o elemento na posição 29+1​=5
, que é o primeiro 12 na distribuição.
Portanto, a mediana das idades é 12.

Exemplo 2 (preços de salgados):

Preços: R$ 3, R$ 3, R$ 5, R$ 6, R$ 6, R$ 8.
Como o número de elementos é par (6), a mediana é a média entre os elementos nas posições 26​=3
 e 26​+1=4
.
A mediana dos preços é (R$ 5 + R$ 6) / 2 = R$ 5,50.

A mediana é útil quando há valores extremos ou distribuição assimétrica.

- Moda

  A moda é o valor numérico que apresenta maior frequência em um conjunto de dados. Ela é útil para diminuir informações de dados com características qualitativas, como categorias ou nomes. Para calcular a moda, siga estes passos:

Organize os dados em ordem crescente ou decrescente.
Verifique o número de ocorrências de cada valor.
O valor com maior frequência é a moda.

- Quartis

  Os quartis são valores que dividem um conjunto de dados ordenados em quatro partes iguais. Eles são usados para entender a dispersão e a distribuição dos dados. Aqui está como calculá-los:


Ordenação dos Dados:

Primeiro, ordene o conjunto de dados em ordem crescente ou decrescente.



Mediana (Segundo Quartil):

A mediana é o valor que divide os dados ao meio.
Se o conjunto tiver número ímpar de elementos, a mediana é a observação central.
Se for número par, a mediana é a média das duas observações centrais.



Primeiro Quartil (Q1):

É a mediana das observações abaixo da mediana.
Se o conjunto tiver número ímpar, Q1 é a observação central das inferiores.
Se for número par, Q1 é a média das duas observações centrais das inferiores.



Terceiro Quartil (Q3):

É a mediana das observações acima da mediana.
Se o conjunto tiver número ímpar, Q3 é a observação central das superiores.
Se for número par, Q3 é a média das duas observações centrais das superiores.



Amplitude Interquartil (AIQ):

É a diferença entre Q3 e Q1: AIQ=Q3−Q1
.



Os quartis são essenciais para análises estatísticas mais avançadas e nos ajudam a compreender a distribuição dos dados1. 

- Desvio Padrão
  O desvio-padrão é uma medida de dispersão utilizada em Estatística para medir o quão os dados se afastam da média aritmética. Ele determina um intervalo onde a maior parte dos dados está concentrada. Quanto maior o valor do desvio-padrão, maior a variabilidade dos dados; quanto menor, menor a irregularidade. Para calcular o desvio-padrão, encontramos a raiz quadrada da variância:
σ=N∑i=1N​(xi​−μ)2​​

(x_1, x_2, x_3, \ldots, x_N) são os dados envolvidos.
(\mu) é a média aritmética dos dados.
(N) é a quantidade de dados.

O desvio-padrão é essencial para entender a variabilidade dos dados e avaliar sua uniformidade.
